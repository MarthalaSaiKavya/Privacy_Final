```python
# ====================================================================================
# BOXRR-23 VR MOTION + DIFFERENTIAL PRIVACY PIPELINE (WINDOW LABELS)
# With EXTRA EXPLAINABILITY + DATA EXPLORATION
#
# What this notebook does (high-level):
# 1) Load a BOXRR-23 VR motion recording (head + hands pose over time).
# 2) Help you UNDERSTAND the dataset using code (EDA + plots).
# 3) Turn raw frames into window-level "high vs low motion" labels.
# 4) Train baseline (non-DP) models and inspect feature importance.
# 5) Apply Differential Privacy:
#       - Laplace & Gaussian noise on features
#       - DP-SGD (Opacus) on an MLP
#       - PATE (teacher ensemble + noisy aggregation)
# 6) Compare all methods and save results to Google Drive.

!pip install -q datasets opacus scikit-learn matplotlib pandas numpy pyarrow

import os
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    confusion_matrix,
)

from opacus import PrivacyEngine
from torch.utils.data import TensorDataset, DataLoader

plt.rcParams["figure.figsize"] = (6, 4)

SEED = 42

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

from google.colab import drive
drive.mount('/content/drive')

SAVE_DIR = "/content/drive/MyDrive/boxrr_dp_project"
os.makedirs(SAVE_DIR, exist_ok=True)
print("Save dir:", SAVE_DIR)
file_url_path = (
    "hf://datasets/cschell/xr-motion-dataset-catalogue/"
    "boxrr23/016e048d-8bac-43a8-bec4-e64799899e13/"
    "00a600eb-ddd9-455c-812f-398b627d3efc.parquet"
)

print("\n=== Loading BOXRR-23 parquet ===")
print("Path:", file_url_path)
df_full = pd.read_parquet(file_url_path)
print("Raw BOXRR shape (frames x columns):", df_full.shape)

print("\nFirst 5 rows (raw frame-level data):")
print(df_full.head())
raw_csv_path = os.path.join(SAVE_DIR, "boxrr23_subset_raw_frames.csv")
df_full.to_csv(raw_csv_path, index=False)
print("\nSaved raw frame-level CSV to:", raw_csv_path)

print("\n=== Basic Info ===")
print(df_full.info())

print("\nColumn names:")
print(list(df_full.columns))

for col in ["user", "session", "app"]:
    if col in df_full.columns:
        print(f"\nUnique values in '{col}':")
        print(df_full[col].value_counts())

print("\n=== Descriptive statistics (numeric columns) ===")
print(df_full.describe().T)

if all(c in df_full.columns for c in ["head_pos_x", "head_pos_y", "head_pos_z"]):
    print("\nPlotting head position over first 300 frames...")
    subset = df_full.iloc[:300]
    plt.figure()
    plt.plot(subset["head_pos_x"], label="head_pos_x")
    plt.plot(subset["head_pos_y"], label="head_pos_y")
    plt.plot(subset["head_pos_z"], label="head_pos_z")
    plt.xlabel("Frame index")
    plt.ylabel("Position")
    plt.title("Head position over time (first 300 frames)")
    plt.legend()
    plt.grid(True)
    plt.show()

drop_cols = [c for c in ["app", "user", "session"] if c in df_full.columns]
feature_cols_frame = [c for c in df_full.columns if c not in drop_cols]
print("\nCandidate frame feature columns (no meta):")
print(feature_cols_frame)


df_frame = df_full[feature_cols_frame].apply(pd.to_numeric, errors="coerce")
df_frame = df_frame.dropna().reset_index(drop=True)

feature_cols_frame = list(df_frame.columns)
print("\nNumeric frame feature columns:")
print(feature_cols_frame)
print("Cleaned frame-level shape:", df_frame.shape)


vec_frame = df_frame[feature_cols_frame].values
motion_norm_frame = np.linalg.norm(vec_frame, axis=1)
df_frame["motion_norm"] = motion_norm_frame

print("\nFirst 5 frame-level motion_norm values:")
print(df_frame[["motion_norm"]].head())

plt.figure()
plt.hist(df_frame["motion_norm"], bins=30)
plt.xlabel("Frame-level motion_norm")
plt.ylabel("Count")
plt.title("Distribution of frame-level motion_norm")
plt.grid(True)
plt.show()


WINDOW = 30

df_frame["window_id"] = df_frame.index // WINDOW

agg_dict = {col: "mean" for col in feature_cols_frame}
agg_dict["motion_norm"] = "mean"

df = df_frame.groupby("window_id").agg(agg_dict).reset_index(drop=True)
print("\nWindow-level shape (windows x features):", df.shape)
print("First 5 window-level rows:")
print(df.head())

median_norm = df["motion_norm"].median()
df["label"] = (df["motion_norm"] > median_norm).astype(int)

print("\nFirst 10 window-level motion_norm + labels:")
print(df[["motion_norm", "label"]].head(10))

print("\nWindow-level class distribution (0=low motion, 1=high motion):")
print(df["label"].value_counts(normalize=True))


plt.figure()
for lbl in [0, 1]:
    subset = df[df["label"] == lbl]
    plt.hist(subset["motion_norm"], bins=15, alpha=0.6, label=f"label = {lbl}")
plt.xlabel("Window-level motion_norm")
plt.ylabel("Count")
plt.title("motion_norm distribution by label")
plt.legend()
plt.grid(True)
plt.show()


df = df.reset_index(drop=True)
df["roll_mean"] = df["motion_norm"].rolling(window=5, min_periods=1).mean()
df["roll_std"]  = df["motion_norm"].rolling(window=5, min_periods=1).std().fillna(0.0)

print("\nFirst 5 rows with temporal features:")
print(df[["motion_norm", "roll_mean", "roll_std", "label"]].head())


plt.figure()
scatter = plt.scatter(df["roll_mean"], df["roll_std"], c=df["label"], cmap="coolwarm", alpha=0.8)
plt.xlabel("roll_mean (recent average motion)")
plt.ylabel("roll_std (recent motion variability)")
plt.title("Window-level temporal features colored by label")
plt.grid(True)
plt.legend(*scatter.legend_elements(), title="label")
plt.show()


base_feature_cols = [c for c in df.columns if c not in ["label"]]
all_features = base_feature_cols

X = df[all_features].values
y = df["label"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)


CLIP_VAL = 3.0
X_train_s = np.clip(X_train_s, -CLIP_VAL, CLIP_VAL)
X_test_s  = np.clip(X_test_s, -CLIP_VAL, CLIP_VAL)

print("\nTrain windows shape:", X_train_s.shape)
print("Test windows shape :", X_test_s.shape)


def eval_and_metrics(clf, X_tr, y_tr, X_te, y_te, name="model", show_cm=False):
    """
    Fit classifier on (X_tr, y_tr), evaluate on (X_te, y_te),
    print accuracy/precision/recall/F1, and optionally confusion matrix.
    """
    clf.fit(X_tr, y_tr)
    y_pred = clf.predict(X_te)
    acc = accuracy_score(y_te, y_pred)
    p, r, f1, _ = precision_recall_fscore_support(
        y_te, y_pred, average="binary", zero_division=0
    )
    print(f"[{name}] acc={acc:.3f}  p={p:.3f}  r={r:.3f}  f1={f1:.3f}")
    if show_cm:
        cm = confusion_matrix(y_te, y_pred)
        print(f"{name} confusion matrix:\n{cm}\n")
    return {"acc": acc, "p": p, "r": r, "f1": f1}

results = []


print("\n=== Baseline models (NON-DP) ===")
baseline_lr = LogisticRegression(max_iter=2000)
metrics_lr = eval_and_metrics(
    baseline_lr, X_train_s, y_train, X_test_s, y_test,
    name="Baseline LR", show_cm=True
)
results.append({
    "name": "Baseline LR",
    "mechanism": "none",
    "eps": None,
    "eps_actual": None,
    "lap_scale": None,
    **metrics_lr
})


coef = baseline_lr.coef_[0]
coef_df = pd.DataFrame({
    "feature": all_features,
    "coef": coef,
    "abs_coef": np.abs(coef),
})
coef_df = coef_df.sort_values("abs_coef", ascending=False)
print("\nTop LR feature importances (by |coefficient|):")
print(coef_df.head(10))


plt.figure()
top_k = 10
plt.barh(coef_df["feature"].head(top_k)[::-1], coef_df["coef"].head(top_k)[::-1])
plt.xlabel("LR coefficient")
plt.title("Top 10 Logistic Regression feature importances")
plt.grid(True, axis="x")
plt.show()


baseline_mlp = MLPClassifier(hidden_layer_sizes=(64, 64), max_iter=500, random_state=SEED)
metrics_mlp = eval_and_metrics(
    baseline_mlp, X_train_s, y_train, X_test_s, y_test,
    name="Baseline MLP (non-DP)", show_cm=True
)
results.append({
    "name": "Baseline MLP (non-DP)",
    "mechanism": "none",
    "eps": None,
    "eps_actual": None,
    "lap_scale": None,
    **metrics_mlp
})


print("\n=== Feature-level DP: Laplace & Gaussian noise ===")

def add_laplace_noise(X, eps, sensitivity=1.0, random_state=None):
    """
    Laplace mechanism: add Laplace noise with scale = sensitivity / eps
    to each feature value.
    Smaller eps = more noise = stronger privacy.
    """
    rng = np.random.default_rng(random_state)
    scale = sensitivity / eps
    noise = rng.laplace(0.0, scale, size=X.shape)
    return X + noise

def add_gaussian_noise(X, eps, delta=1e-5, sensitivity=1.0, random_state=None):
    """
    Gaussian mechanism: add Gaussian noise with std = sigma(eps, delta)
    to each feature.
    """
    rng = np.random.default_rng(random_state)
    sigma = math.sqrt(2 * math.log(1.25 / delta)) * sensitivity / eps
    noise = rng.normal(0.0, sigma, size=X.shape)
    return X + noise

eps_list = [0.1, 0.5, 1.0, 2.0, 5.0]

laplace_results = []
gauss_results   = []

for eps in eps_list:

    X_train_lap = add_laplace_noise(X_train_s, eps, sensitivity=1.0, random_state=SEED)
    clf_lap = LogisticRegression(max_iter=2000)
    lap_metrics = eval_and_metrics(
        clf_lap, X_train_lap, y_train, X_test_s, y_test,
        name=f"LR + Laplace eps={eps}", show_cm=False
    )
    lap_res = {
        "name": f"LR + Laplace eps={eps}",
        "mechanism": "laplace",
        "eps": eps,
        "eps_actual": None,
        "lap_scale": None,
        **lap_metrics
    }
    laplace_results.append(lap_res)
    results.append(lap_res)


    X_train_gauss = add_gaussian_noise(
        X_train_s, eps, delta=1e-5, sensitivity=1.0, random_state=SEED
    )
    clf_gauss = LogisticRegression(max_iter=2000)
    ga_metrics = eval_and_metrics(
        clf_gauss, X_train_gauss, y_train, X_test_s, y_test,
        name=f"LR + Gaussian eps={eps}", show_cm=False
    )
    ga_res = {
        "name": f"LR + Gaussian eps={eps}",
        "mechanism": "gaussian",
        "eps": eps,
        "eps_actual": None,
        "lap_scale": None,
        **ga_metrics
    }
    gauss_results.append(ga_res)
    results.append(ga_res)

def extract_acc(results_list):
    return [m["acc"] for m in results_list]

plt.figure()
plt.plot(eps_list, extract_acc(laplace_results), marker='o', label="Laplace DP LR")
plt.plot(eps_list, extract_acc(gauss_results), marker='s', label="Gaussian DP LR")
plt.axhline(metrics_lr["acc"], linestyle='--', label="Baseline LR")
plt.xscale("log")
plt.xlabel("epsilon (log scale, smaller = more private)")
plt.ylabel("Test Accuracy")
plt.title("LR Accuracy vs epsilon (Laplace vs Gaussian)")
plt.grid(True)
plt.legend()
plt.show()


print("\n=== DP-SGD on MLP (Opacus) ===")
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

train_t = torch.tensor(X_train_s, dtype=torch.float32)
label_t = torch.tensor(y_train, dtype=torch.long)

train_ds = TensorDataset(train_t, label_t)
train_dl = DataLoader(train_ds, batch_size=8, shuffle=True)
input_dim = X_train_s.shape[1]

class MLP(nn.Module):
    def __init__(self, in_dim, hid=64, out=2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hid),
            nn.ReLU(),
            nn.Linear(hid, hid),
            nn.ReLU(),
            nn.Linear(hid, out),
        )
    def forward(self, x):
        return self.net(x)

def train_dp_sgd(target_eps=3.0, delta=1e-5, epochs=5, max_grad_norm=1.0):
    """
    Train an MLP with DP-SGD using Opacus:
      - Clipping gradients to max_grad_norm
      - Adding Gaussian noise to gradients
      - Target epsilon at the end of training
    """
    model = MLP(input_dim).to(device)
    optimizer = optim.Adam(model.parameters(), lr=5e-4)
    criterion = nn.CrossEntropyLoss()

    privacy_engine = PrivacyEngine()
    model, optimizer, private_dl = privacy_engine.make_private_with_epsilon(
        module=model,
        optimizer=optimizer,
        data_loader=train_dl,
        target_epsilon=target_eps,
        target_delta=delta,
        max_grad_norm=max_grad_norm,
        epochs=epochs,
    )

    for epoch in range(epochs):
        model.train()
        total_loss = 0.0
        for xb, yb in private_dl:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            logits = model(xb)
            loss = criterion(logits, yb)
            loss.backward()
            optimizer.step()
            total_loss += loss.item() * xb.size(0)
        print(f"[DP-SGD] eps_target={target_eps} Epoch {epoch+1}/{epochs}, loss={total_loss/len(private_dl.dataset):.4f}")

    eps = privacy_engine.get_epsilon(delta)
    print(f"Achieved (eps, delta)=({eps:.2f}, {delta}) for target_eps={target_eps}")
    return model, eps

eps_values_dp = [1.0, 3.0, 5.0]
dp_sgd_results = []

for eps_target in eps_values_dp:
    print("\n=== DP-SGD run, target_eps =", eps_target, "===")
    dp_model, eps_actual = train_dp_sgd(target_eps=eps_target, delta=1e-5, epochs=5)

    dp_model.eval()
    with torch.no_grad():
        X_test_t = torch.tensor(X_test_s, dtype=torch.float32).to(device)
        logits_dp = dp_model(X_test_t)
        y_pred_dp = logits_dp.argmax(1).cpu().numpy()

    acc_dp = accuracy_score(y_test, y_pred_dp)
    p_dp, r_dp, f1_dp, _ = precision_recall_fscore_support(
        y_test, y_pred_dp, average="binary", zero_division=0
    )
    res = {
        "name": f"DP-SGD MLP target_eps={eps_target:.1f} (actual≈{float(eps_actual):.2f})",
        "mechanism": "dp-sgd",
        "eps": eps_target,
        "eps_actual": float(eps_actual),
        "lap_scale": None,
        "acc": acc_dp,
        "p": p_dp,
        "r": r_dp,
        "f1": f1_dp,
    }
    dp_sgd_results.append(res)
    results.append(res)
    print(res)

plt.figure()
plt.plot([r["eps_actual"] for r in dp_sgd_results],
         [r["acc"] for r in dp_sgd_results],
         marker='o', label="DP-SGD MLP")
plt.axhline(metrics_mlp["acc"], linestyle='--', label="Non-DP MLP baseline")
plt.xlabel("actual epsilon (DP-SGD)")
plt.ylabel("Test Accuracy")
plt.title("DP-SGD MLP Accuracy vs epsilon")
plt.grid(True)
plt.legend()
plt.show()


print("\n=== PATE (Private Aggregation of Teacher Ensembles) ===")


X_priv, X_pub, y_priv, y_pub = train_test_split(
    X_train_s, y_train, test_size=0.3, random_state=SEED, stratify=y_train
)

def build_teachers(X, y, num_teachers=4):
    """
    Trains 'num_teachers' LogisticRegression models on disjoint slices of X,y.
    Each teacher sees a different subset of the private data.
    """
    idx = np.random.permutation(len(X))
    splits = np.array_split(idx, num_teachers)
    teachers = []
    for i, s in enumerate(splits):
        clf = LogisticRegression(max_iter=2000)
        clf.fit(X[s], y[s])
        teachers.append(clf)
        print(f"Trained PATE teacher {i+1}/{num_teachers} on {len(s)} samples.")
    return teachers

teachers = build_teachers(X_priv, y_priv, num_teachers=4)

def pate_noisy_labels(teachers, X_pool, lap_scale=1.0):
    """
    For each example in X_pool:
     - Each teacher predicts a label (0 or 1).
     - We count votes across teachers.
     - We add Laplace noise to the vote counts.
     - The final label is the argmax of the NOISY vote counts.
    """
    votes = np.zeros((len(X_pool), 2), dtype=np.float64)  # 2 classes
    for t in teachers:
        preds = t.predict(X_pool)
        for i, c in enumerate(preds):
            votes[i, int(c)] += 1.0


    noisy_votes = votes + np.random.laplace(0, lap_scale, votes.shape)
    labels = noisy_votes.argmax(axis=1)
    return labels, votes, noisy_votes

lap_scales = [0.3, 1.0, 3.0]
pate_results = []

for s in lap_scales:
    print("\n=== PATE with Laplace scale =", s, "===")

    pate_labels, raw_votes, noisy_votes = pate_noisy_labels(teachers, X_pub, lap_scale=s)
    print("First 5 raw votes (before noise):\n", raw_votes[:5])
    print("First 5 noisy votes (after noise):\n", noisy_votes[:5])
    print("First 10 PATE labels:", pate_labels[:10])

    student = LogisticRegression(max_iter=2000)
    student.fit(X_pub, pate_labels)

    y_pred_stud = student.predict(X_test_s)
    acc_stud = accuracy_score(y_test, y_pred_stud)
    p_stud, r_stud, f1_stud, _ = precision_recall_fscore_support(
        y_test, y_pred_stud, average="binary", zero_division=0
    )
    res = {
        "name": f"PATE Student LR (lap_scale={s})",
        "mechanism": "pate",
        "eps": None,
        "eps_actual": None,
        "lap_scale": s,
        "acc": acc_stud,
        "p": p_stud,
        "r": r_stud,
        "f1": f1_stud,
    }
    pate_results.append(res)
    results.append(res)
    print(res)


plt.figure()
plt.plot(lap_scales, [r["acc"] for r in pate_results], marker='o')
plt.xlabel("Laplace noise scale (vote perturbation)")
plt.ylabel("PATE Student Accuracy")
plt.title("PATE Accuracy vs noise scale")
plt.grid(True)
plt.show()



import torch
import torch.nn as nn
import torch.optim as optim
from opacus import PrivacyEngine
from torch.utils.data import DataLoader
from sklearn.linear_model import LogisticRegression

class Generator(nn.Module):
    def __init__(self, noise_dim: int, data_dim: int, hidden_dim: int = 64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(noise_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, data_dim),
        )

    def forward(self, z):
        return self.net(z)


class Discriminator(nn.Module):
    def __init__(self, data_dim: int, hidden_dim: int = 64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(data_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1),
            nn.Sigmoid(),
        )

    def forward(self, x):
        return self.net(x)



def train_pategan_gan(
    X_train,
    noise_dim: int = 16,
    epochs: int = 50,
    batch_size: int = 128,
    lr: float = 1e-3,
    device: str = "cpu",
):
    """
    Train a *vanilla* GAN (Generator + Discriminator) on X_train.
    This is the backbone for PATE-GAN: the *labels* for synthetic data
    will come from PATE (teachers + noisy aggregation), not from the GAN.

    Returns:
        Trained Generator (G)
    """
    X_train_t = torch.tensor(X_train, dtype=torch.float32).to(device)
    data_dim = X_train.shape[1]

    G = Generator(noise_dim=noise_dim, data_dim=data_dim).to(device)
    D = Discriminator(data_dim=data_dim).to(device)

    opt_G = optim.Adam(G.parameters(), lr=lr)
    opt_D = optim.Adam(D.parameters(), lr=lr)

    bce = nn.BCELoss()
    loader = DataLoader(X_train_t, batch_size=batch_size, shuffle=True)

    print("\n=== Training PATE-GAN (vanilla GAN backbone) ===")
    for epoch in range(1, epochs + 1):
        for x_real in loader:
            x_real = x_real.to(device)
            bs = x_real.size(0)


            opt_D.zero_grad()
            y_real = torch.ones(bs, 1, device=device)
            y_fake = torch.zeros(bs, 1, device=device)


            d_real = D(x_real)
            loss_real = bce(d_real, y_real)


            z = torch.randn(bs, noise_dim, device=device)
            x_fake = G(z).detach()
            d_fake = D(x_fake)
            loss_fake = bce(d_fake, y_fake)

            loss_D = loss_real + loss_fake
            loss_D.backward()
            opt_D.step()


            opt_G.zero_grad()
            z = torch.randn(bs, noise_dim, device=device)
            x_fake = G(z)
            d_fake_for_G = D(x_fake)
            loss_G = bce(d_fake_for_G, y_real)
            loss_G.backward()
            opt_G.step()

        if epoch == 1 or epoch % 10 == 0:
            print(
                f"[PATE-GAN] Epoch {epoch}/{epochs} "
                f"loss_D={loss_D.item():.4f}  loss_G={loss_G.item():.4f}"
            )

    return G



pategan_lap_scales = [0.3, 1.0, 3.0]
pategan_results = []


G_pategan = train_pategan_gan(
    X_train_s,
    noise_dim=16,
    epochs=50,
    batch_size=128,
    lr=1e-3,
    device=device,
)


n_synth = X_train_s.shape[0]

z_base = torch.randn(n_synth, 16, device=device)

for s in pategan_lap_scales:
    print(f"\n=== PATE-GAN with Laplace scale = {s} ===")


    with torch.no_grad():
        X_synth = G_pategan(z_base).cpu().numpy()


    pate_labels_synth, raw_votes_synth, noisy_votes_synth = pate_noisy_labels(
        teachers,
        X_synth,
        lap_scale=s,
    )

    print("First 5 PATE-GAN raw votes:\n", raw_votes_synth[:5])
    print("First 5 PATE-GAN noisy votes:\n", noisy_votes_synth[:5])
    print("First 10 PATE-GAN labels:", pate_labels_synth[:10])


    clf_pategan = LogisticRegression(max_iter=2000)
    metr_pg = eval_and_metrics(
        clf_pategan,
        X_synth,
        pate_labels_synth,
        X_test_s,
        y_test,
        name=f"PATE-GAN LR (lap_scale={s})",
        show_cm=False,
    )


    res_pg = {
        "name": f"PATE-GAN LR (lap_scale={s})",
        "mechanism": "pate-gan",
        "eps": None,
        "eps_actual": None,
        "lap_scale": s,
        "acc": metr_pg["acc"],
        "p": metr_pg["p"],
        "r": metr_pg["r"],
        "f1": metr_pg["f1"],
    }
    pategan_results.append(res_pg)
    results.append(res_pg)


pategan_df = pd.DataFrame(pategan_results)
print("\n=== PATE-GAN SUMMARY ===")
print(pategan_df[["lap_scale", "acc", "p", "r", "f1"]].round(3))



pate_df = pd.DataFrame(pate_results)
pategan_df = pd.DataFrame(pategan_results)


pate_df = pate_df.sort_values("lap_scale")
pategan_df = pategan_df.sort_values("lap_scale")

plt.figure()
plt.plot(
    pate_df["lap_scale"],
    pate_df["acc"],
    marker="o",
    label="PATE Student LR",
)
plt.plot(
    pategan_df["lap_scale"],
    pategan_df["acc"],
    marker="s",
    label="PATE-GAN LR",
)
plt.xlabel("Laplace noise scale (vote perturbation)")
plt.ylabel("Test Accuracy")
plt.title("Accuracy vs noise scale: PATE vs PATE-GAN")
plt.grid(True)
plt.legend()
plt.show()

def run_dpgan_experiment(
    X_train,
    y_train,
    X_test,
    y_test,
    target_eps: float = 1.0,
    delta: float = 1e-5,
    batch_size: int = 128,
    max_grad_norm: float = 1.0,
    epochs: int = 50,
    noise_dim: int = 16,
    lr: float = 1e-3,
    device: str = "cpu",
    noise_multiplier: float = None,
):
    """
    1) Train a DP-GAN on X_train (labels not used in GAN itself).
    2) Generate synthetic samples from the generator.
    3) Train Logistic Regression on synthetic data, evaluate on real test data.

    Returns:
        eps_actual (float or None),
        metrics (dict with acc, p, r, f1)
    """
    X_train_t = torch.tensor(X_train, dtype=torch.float32).to(device)
    data_dim = X_train.shape[1]

    G = Generator(noise_dim=noise_dim, data_dim=data_dim).to(device)
    D = Discriminator(data_dim=data_dim).to(device)

    opt_G = optim.Adam(G.parameters(), lr=lr)
    opt_D = optim.Adam(D.parameters(), lr=lr)


    train_loader = DataLoader(X_train_t, batch_size=batch_size, shuffle=True)


    if noise_multiplier is None:
        if target_eps <= 1.0:
            noise_multiplier = 2.0
        elif target_eps <= 3.0:
            noise_multiplier = 1.0
        else:
            noise_multiplier = 0.5

    privacy_engine = PrivacyEngine()
    D, opt_D, train_loader = privacy_engine.make_private(
        module=D,
        optimizer=opt_D,
        data_loader=train_loader,
        noise_multiplier=noise_multiplier,
        max_grad_norm=max_grad_norm,
        poisson_sampling=False,
    )

    bce = nn.BCELoss()

    print(f"\n=== DP-GAN run: target_eps={target_eps}, noise_multiplier={noise_multiplier} ===")
    for epoch in range(1, epochs + 1):
        D.train()
        G.train()

        for x_real in train_loader:
            x_real = x_real.to(device)
            bs = x_real.size(0)


            opt_D.zero_grad()

            y_real = torch.ones(bs, 1, device=device)
            y_fake = torch.zeros(bs, 1, device=device)


            d_real = D(x_real)
            loss_real = bce(d_real, y_real)


            z = torch.randn(bs, noise_dim, device=device)
            x_fake = G(z).detach()
            d_fake = D(x_fake)
            loss_fake = bce(d_fake, y_fake)

            loss_D = loss_real + loss_fake
            loss_D.backward()
            opt_D.step()

            opt_G.zero_grad()
            z = torch.randn(bs, noise_dim, device=device)
            x_fake = G(z)
            d_fake_for_G = D(x_fake)
            loss_G = bce(d_fake_for_G, y_real)
            loss_G.backward()
            opt_G.step()

        if epoch == 1 or epoch % 10 == 0:
            print(f"[DP-GAN] Epoch {epoch}/{epochs} done")


    try:
        eps_actual = privacy_engine.get_epsilon(delta=delta)
    except Exception:
        eps_actual = None

    print(
        f"[DP-GAN] Achieved eps (delta={delta}): "
        f"{eps_actual if eps_actual is not None else 'N/A'} "
        f"for target_eps={target_eps}"
    )


    G.eval()
    n_synth = X_train.shape[0]
    z = torch.randn(n_synth, noise_dim, device=device)
    with torch.no_grad():
        X_synth = G(z).cpu().numpy()


    y_synth = y_train


    clf = LogisticRegression(max_iter=2000)
    metrics = eval_and_metrics(
        clf,
        X_synth,
        y_synth,
        X_test,
        y_test,
        name=f"DP-GAN LR eps_target={target_eps}",
        show_cm=False,
    )

    print(
        f"[DP-GAN] Metrics: acc={metrics['acc']:.3f}  "
        f"p={metrics['p']:.3f}  r={metrics['r']:.3f}  f1={metrics['f1']:.3f}"
    )

    return eps_actual, metrics



dpgan_eps_targets = [1.0, 3.0, 5.0]
delta = 1e-5
dpgan_results = []

for tgt in dpgan_eps_targets:
    eps_act, metr = run_dpgan_experiment(
        X_train=X_train_s,
        y_train=y_train,
        X_test=X_test_s,
        y_test=y_test,
        target_eps=tgt,
        delta=delta,
        batch_size=128,
        max_grad_norm=1.0,
        epochs=50,
        noise_dim=16,
        lr=1e-3,
        device=device,
    )


    if eps_act is not None:
        eps_display = f"{float(eps_act):.2f}"
    else:
        eps_display = "nan"

    res = {
        "name": f"DP-GAN LR target_eps={tgt} (actual≈{eps_display})",
        "mechanism": "dp-gan",
        "eps": tgt,
        "eps_actual": float(eps_act) if eps_act is not None else None,
        "lap_scale": None,
        "acc": metr["acc"],
        "p": metr["p"],
        "r": metr["r"],
        "f1": metr["f1"],
    }
    dpgan_results.append(res)
    results.append(res)


dpgan_df = pd.DataFrame(dpgan_results)
print("\n=== DP-GAN SUMMARY (rounded) ===")
print(
    dpgan_df[["eps", "eps_actual", "acc", "p", "r", "f1"]]
    .sort_values("eps")
    .round(3)
)

results_df = pd.DataFrame(results)


dpgan_df = results_df[results_df["mechanism"] == "dp-gan"].copy()
dpgan_df = dpgan_df.sort_values("eps_actual")

plt.figure()
plt.plot(dpgan_df["eps_actual"], dpgan_df["acc"], marker="o", label="DP-GAN LR")
plt.xlabel("actual epsilon (DP-GAN)")
plt.ylabel("Test Accuracy")
plt.title("DP-GAN LR Accuracy vs epsilon")
plt.grid(True)
plt.legend()
plt.show()


dp_sgd_df = results_df[results_df["mechanism"] == "dp-sgd"].copy()
dp_sgd_df = dp_sgd_df.sort_values("eps_actual")

plt.figure()
plt.plot(dp_sgd_df["eps_actual"], dp_sgd_df["acc"],
         marker="o", label="DP-SGD MLP")
plt.plot(dpgan_df["eps_actual"], dpgan_df["acc"],
         marker="s", label="DP-GAN LR")
plt.xlabel("actual epsilon")
plt.ylabel("Test Accuracy")
plt.title("Accuracy vs ε: DP-SGD vs DP-GAN")
plt.grid(True)
plt.legend()
plt.show()


print("\n=== FINAL RESULTS TABLE ===")
print(results_df)

results_path = os.path.join(SAVE_DIR, "boxrr23_dp_results_windows.csv")
os.makedirs(os.path.dirname(results_path), exist_ok=True)
results_df.to_csv(results_path, index=False)
print("\nSaved DP window-level results to:", results_path)
```
